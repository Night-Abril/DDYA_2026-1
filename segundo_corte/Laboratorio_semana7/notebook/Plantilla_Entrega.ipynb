## 1) Descripción del problema
Este laboratorio consiste en la implementación de estructuras de datos fundamentales como pilas y colas utilizando Python. Estas estructuras se aplican en la simulación de un sistema de atención médica donde los pacientes son atendidos según su prioridad.

En el sistema existen dos tipos de pacientes: pacientes normales y pacientes críticos. Los pacientes normales son atendidos siguiendo el orden de llegada, mientras que los pacientes críticos tienen prioridad sobre los normales. Sin embargo, entre pacientes críticos también se respeta el orden en que llegaron.

El sistema recibe una serie de acciones que representan la llegada de pacientes o la atención de uno de ellos. El objetivo es gestionar estas acciones utilizando estructuras de datos adecuadas para garantizar que las reglas del sistema se cumplan correctamente.

## 2) Requerimientos
Lista requerimientos **funcionales** 
- El sistema debe permitir registrar pacientes normales.

- El sistema debe permitir registrar pacientes críticos.

- El sistema debe atender pacientes según su prioridad.

- El sistema debe mostrar qué paciente fue atendido.

- El sistema debe indicar cuando no hay pacientes disponibles.

Lista requerimientos **no funcionales**
- El sistema debe estar implementado en Python.

- No se deben utilizar librerías externas como collections o deque.

- El código debe ser claro y modular.

- El programa debe ejecutarse desde consola.

## 4) Diagramas de flujo
El diagrama de flujo describe cómo funciona el sistema de atención médica que se ha implementado en el programa. Todo comienza cuando se le pide al usuario que indique cuántas acciones llevará a cabo. Luego, el sistema entra en un ciclo en el que se leen las acciones que el usuario ha introducido.

Según el tipo de acción, el sistema puede registrar a un paciente normal, a uno crítico o atender a un paciente. Al registrar un paciente, se lo coloca en la cola correspondiente según su prioridad. Si la acción consiste en atender, el sistema primero verifica si hay pacientes críticos, dándoles atención antes que a los normales. En caso de que no haya pacientes en ninguna cola, el sistema avisa que no hay pacientes disponibles.

Este procedimiento se repite hasta que se haya cumplido con el número de acciones que se estableció al principio. Una vez que todas las acciones han sido procesadas, el sistema concluye.


## 5) Diagramas de secuencia
El diagrama de secuencia muestra cómo interactúan el recepcionista, el sistema y las dos colas de pacientes. Cuando el usuario registra un paciente, el sistema lo agrega a la cola correspondiente según su prioridad. Cuando se solicita atender un paciente, el sistema primero verifica la cola de pacientes críticos. Si existe un paciente crítico, se atiende primero. Si no hay críticos, se revisa la cola de pacientes normales. Si ambas colas están vacías, el sistema indica que no hay pacientes.

## 7) Análisis de complejidad
El sistema utiliza estructuras de cola para gestionar la atención de pacientes. Las operaciones principales son encolar y desencolar pacientes.

- La operación encolar tiene una complejidad de tiempo O(1).

- La operación desencolar también tiene una complejidad O(1).

- La consulta del primer elemento de la cola tiene complejidad O(1).

Por lo tanto, el sistema mantiene una eficiencia constante para las operaciones principales, incluso cuando el número de pacientes aumenta.

## 9) Código funcional



```python
class Cola:
    def __init__(self):
        self.elementos = []

    def encolar(self, elemento):
        self.elementos.append(elemento)

    def desencolar(self):
        if not self.esta_vacia():
            return self.elementos.pop(0)
        return None

    def frente(self):
        if not self.esta_vacia():
            return self.elementos[0]
        return None

    def esta_vacia(self):
        return len(self.elementos) == 0

    def tamaño(self):
        return len(self.elementos)

def sistema_pacientes():
    cola_criticos = Cola()
    cola_normales = Cola()
    print("Sistema de atención médica")
    print("Ingrese el número de acciones:")
    n = int(input())
    print("Ingrese las acciones (N nombre, C nombre o A):")

    for _ in range(n):
        accion = input().split()

        if accion[0] == "N":
            nombre = accion[1]
            cola_normales.encolar(nombre)
            print("Paciente N nuevo :", nombre)

        elif accion[0] == "C":
            nombre = accion[1]
            cola_criticos.encolar(nombre)
            print("Paciente C nuevo :", nombre)

        elif accion[0] == "A":
            if not cola_criticos.esta_vacia():
                paciente = cola_criticos.desencolar()
                print("Se atiende a", paciente)

            elif not cola_normales.esta_vacia():
                paciente = cola_normales.desencolar()
                print("Se atiende a", paciente)

            else:
                print("no hay pacientes")

if __name__ == "__main__":
    sistema_pacientes()
```
